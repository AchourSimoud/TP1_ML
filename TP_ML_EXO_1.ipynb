```python
data=pd.read_csv('data/mouse.txt',sep=' ',header=None)
data.head(10)
data_mat=data.to_numpy()
```

```python
fig, ax = plt.subplots()
ax.scatter(data_mat[:,0], data_mat[:,1])
plt.show
```

```python
IF=IsolationForest(n_estimators=500,random_state=1)
IF.fit(data)
```

```python
dec=IF.decision_function(data)
fig, ax = plt.subplots()
ax.hist(dec,linewidth=0.5, edgecolor="white", bins=50)
plt.show
```

```python
outlier=dec.copy()
outlier[dec<=-0.10]=1
outlier[dec>-0.10]=0
fig, ax = plt.subplots()
ax.scatter(data_mat[:,0], data_mat[:,1], c=outlier)
plt.show
```

```python
LOF=LocalOutlierFactor(n_neighbors=35)
prediction=LOF.fit(data)
```

```python
LOF_dec=LOF.negative_outlier_factor_
fig, ax = plt.subplots()
ax.hist(LOF_dec,linewidth=0.5, edgecolor="white", bins=50)
plt.show
```

```python
outlier=LOF_dec.copy()
outlier[LOF_dec<=-1.5]=1
outlier[LOF_dec>-1.5]=0
fig, ax = plt.subplots()
ax.scatter(data_mat[:,0], data_mat[:,1], c=outlier)
plt.show
```
